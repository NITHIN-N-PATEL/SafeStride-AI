# P3: Staircase Preprocessing

**Source:** `Dataset/Staircase/`

**Requirements:**
1. **Merge** 'Upstair' and 'Downstair' into a single **Stairs** class.
2. **Remap to ID 83**.
3. Output to `processed_data/p3_staircase/`.

```python
import os
import shutil
from pathlib import Path
from collections import Counter

# --- CONFIG ---
BASE_DIR   = Path(r"c:\SafeStride\Dataset preprocessing")
INPUT_DIR  = BASE_DIR / "Dataset" / "Staircase"
OUTPUT_DIR = BASE_DIR / "processed_data" / "p3_staircase"

images_out = OUTPUT_DIR / "images"
labels_out = OUTPUT_DIR / "labels"

# Merge both classes into 83
P3_REMAP = {0: 83, 1: 83, 2: 83} # Added 2 just in case of 1-indexing

IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp"}

print(f"Input: {INPUT_DIR}")
print(f"Output: {OUTPUT_DIR}")
```

## 1 — Verify Class IDs in Source

```python
counts_raw = Counter()
for lf in INPUT_DIR.rglob("*.txt"):
    if "README" in lf.name: continue
    for line in lf.read_text().splitlines():
        parts = line.split()
        if parts: counts_raw[parts[0]] += 1

print("Raw IDs found in Staircase dataset:")
for cid, count in sorted(counts_raw.items()):
    print(f"  ID {cid}: {count} instances")
```

## 2 — Run Processing

```python
if OUTPUT_DIR.exists():
    shutil.rmtree(OUTPUT_DIR)
images_out.mkdir(parents=True, exist_ok=True)
labels_out.mkdir(parents=True, exist_ok=True)

copied = 0
lines_kept = 0

for split in ["train", "test", "valid"]:
    img_in = INPUT_DIR / split / "images"
    lbl_in = INPUT_DIR / split / "labels"
    
    if not lbl_in.exists(): continue
    
    for lf in lbl_in.glob("*.txt"):
        new_lines = []
        for line in lf.read_text().splitlines():
            parts = line.strip().split()
            if not parts: continue
            
            cid = int(parts[0])
            if cid in P3_REMAP:
                parts[0] = str(P3_REMAP[cid])
                new_lines.append(" ".join(parts))
                lines_kept += 1
        
        if not new_lines: continue
        
        img_src = None
        for ext in IMG_EXTS:
            candidate = img_in / f"{lf.stem}{ext}"
            if candidate.exists():
                img_src = candidate
                break
        
        if img_src:
            shutil.copy2(img_src, images_out / img_src.name)
            (labels_out / lf.name).write_text("\n".join(new_lines) + "\n")
            copied += 1

print(f"Finished P3 Processing!")
print(f"  Pairs saved: {copied}")
print(f"  Total 'Stairs' annotations: {lines_kept}")
```
