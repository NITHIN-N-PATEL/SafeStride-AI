```python
import os
import shutil
from pathlib import Path
from collections import Counter

# --- CONFIG ---
BASE_DIR   = Path(r"c:\SafeStride\Dataset preprocessing")
INPUT_DIR  = BASE_DIR / "Dataset" / "Manhole"
OUTPUT_DIR = BASE_DIR / "processed_data" / "p2_manhole"

images_out = OUTPUT_DIR / "images"
labels_out = OUTPUT_DIR / "labels"

# Diagnostic check showed that ID 3 is the only one present in this dataset export.
KEEP_ID = 3  
TARGET_ID = 82

IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp"}

print(f"Input: {INPUT_DIR}")
print(f"Output: {OUTPUT_DIR}")
print(f"Filtering for Source ID: {KEEP_ID} -> Target ID: {TARGET_ID}")
```

## 1 — Verify Class IDs in Source
Run this to see which IDs are actually present in your Manhole folders.

```python
counts_raw = Counter()
for lf in INPUT_DIR.rglob("*.txt"):
    if "README" in lf.name: continue
    for line in lf.read_text().splitlines():
        parts = line.split()
        if parts: counts_raw[parts[0]] += 1

print("Raw IDs found in Manhole dataset:")
for cid, count in sorted(counts_raw.items()):
    print(f"  ID {cid}: {count} instances")
```

## 2 — Run Processing

```python
if OUTPUT_DIR.exists():
    shutil.rmtree(OUTPUT_DIR)
images_out.mkdir(parents=True, exist_ok=True)
labels_out.mkdir(parents=True, exist_ok=True)

copied = 0
lines_kept = 0

# Process all subfolders (train, test, valid)
for split in ["train", "test", "valid"]:
    img_in = INPUT_DIR / split / "images"
    lbl_in = INPUT_DIR / split / "labels"
    
    if not lbl_in.exists(): continue
    
    for lf in lbl_in.glob("*.txt"):
        new_lines = []
        for line in lf.read_text().splitlines():
            parts = line.strip().split()
            if not parts: continue
            
            cid = int(parts[0])
            # Only keep the specific ID we want
            if cid == KEEP_ID:
                parts[0] = str(TARGET_ID)
                new_lines.append(" ".join(parts))
                lines_kept += 1
        
        if not new_lines: continue
        
        # Find matching image
        img_src = None
        for ext in IMG_EXTS:
            candidate = img_in / f"{lf.stem}{ext}"
            if candidate.exists():
                img_src = candidate
                break
        
        if img_src:
            shutil.copy2(img_src, images_out / img_src.name)
            (labels_out / lf.name).write_text("\n".join(new_lines) + "\n")
            copied += 1

print(f"Finished P2 Processing!")
print(f"  Pairs saved: {copied}")
print(f"  Total 'Open Manhole' annotations kept: {lines_kept}")
if copied == 0:
    print("[!] WARNING: No pairs were saved. Please check if 'KEEP_ID' matches your dataset IDs.")
```
