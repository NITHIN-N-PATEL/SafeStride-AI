# P1: Pothole & Road Crack Preprocessing

**Source:** `Dataset/rdd-pothole/p1_dataset/`

**Goals:**
1. Convert OBB (9-value) labels to standard YOLO (5-value) bboxes.
2. Filter classes: Keep only Pothole and Road Crack.
3. **Remap IDs to 80 and 81** to avoid collision with COCO (0-79).
4. Output to `processed_data/p1_pothole/`.

```python
import os
import shutil
from pathlib import Path
from collections import Counter

BASE_DIR   = Path(r"c:\SafeStride\Dataset preprocessing")
INPUT_DIR  = BASE_DIR / "Dataset" / "rdd-pothole" / "p1_dataset"
OUTPUT_DIR = BASE_DIR / "processed_data" / "p1_pothole"

images_in  = INPUT_DIR / "images"
labels_in  = INPUT_DIR / "labels"

images_out = OUTPUT_DIR / "images"
labels_out = OUTPUT_DIR / "labels"

# Global Mappings
KEEP_CLASSES = {0, 1} 
P1_REMAP     = {0: 80, 1: 81} # 80: Pothole, 81: Road Crack

IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp"}

print(f"Input: {INPUT_DIR}")
print(f"Output: {OUTPUT_DIR}")
```

## 1 — Helper Functions

```python
def obb_to_yolo_bbox(parts):
    """
    Convert a 9-value OBB line to 5-value standard YOLO bbox.
    Input:  ['class', 'x1', 'y1', 'x2', 'y2', 'x3', 'y3', 'x4', 'y4']
    Output: 'class x_center y_center width height'
    """
    class_id = parts[0]
    coords = [float(v) for v in parts[1:]]
    xs = coords[0::2]  # x1, x2, x3, x4
    ys = coords[1::2]  # y1, y2, y3, y4
    
    x_min, x_max = min(xs), max(xs)
    y_min, y_max = min(ys), max(ys)
    
    x_center = (x_min + x_max) / 2
    y_center = (y_min + y_max) / 2
    width    = x_max - x_min
    height   = y_max - y_min
    
    return f"{class_id} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}"
```

## 2 — Run Processing

```python
if OUTPUT_DIR.exists():
    shutil.rmtree(OUTPUT_DIR)
images_out.mkdir(parents=True, exist_ok=True)
labels_out.mkdir(parents=True, exist_ok=True)

copied = 0
lines_kept = 0
counts_after = Counter()

label_files = sorted(list(labels_in.glob("*.txt")))

for lf in label_files:
    original_lines = lf.read_text().splitlines()
    new_lines = []
    
    for line in original_lines:
        parts = line.strip().split()
        if not parts: continue
        
        class_id = int(parts[0])
        if class_id not in KEEP_CLASSES: continue
        
        # Remap ID
        new_id = P1_REMAP[class_id]
        parts[0] = str(new_id)
        
        if len(parts) == 9:
            new_lines.append(obb_to_yolo_bbox(parts))
            lines_kept += 1
            counts_after[new_id] += 1
        elif len(parts) == 5:
            new_lines.append(" ".join(parts))
            lines_kept += 1
            counts_after[new_id] += 1
            
    if not new_lines: continue
    
    # Match Image
    img_src = None
    for ext in IMG_EXTS:
        candidate = images_in / f"{lf.stem}{ext}"
        if candidate.exists():
            img_src = candidate
            break
    
    if img_src:
        shutil.copy2(img_src, images_out / img_src.name)
        (labels_out / lf.name).write_text("\n".join(new_lines) + "\n")
        copied += 1

print(f"Finished P1 Processing!")
print(f"  Pairs saved: {copied}")
print(f"  Total annotations: {lines_kept}")
for cid, count in counts_after.items():
    name = "Pothole" if cid == 80 else "Road Crack"
    print(f"    ID {cid} ({name}): {count}")
```
