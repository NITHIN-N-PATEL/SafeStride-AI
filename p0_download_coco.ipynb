# P0: Download COCO Reminder Set

This notebook uses **FiftyOne** to download the COCO 2017 Validation dataset (5,000 images) and export it in YOLO format for use in our merging pipeline.

```python
# Install FiftyOne if not already installed
%pip install fiftyone
```

```python
import shutil
from pathlib import Path

# --- CONFIG ---
EXPORT_DIR = "coco_reminder_set"
coco_path = Path(EXPORT_DIR)

print("Finishing flattening folder structure...")

for folder in ["images", "labels"]:
    val_path = coco_path / folder / "val"
    
    if val_path.exists():
        # 1. Move files if any are left
        files = list(val_path.iterdir())
        if files:
            print(f"Moving {len(files)} files from {val_path}...")
            for file in files:
                shutil.move(str(file), str(coco_path / folder / file.name))
        
        # 2. Try to delete the empty val folder
        try:
            val_path.rmdir()
            print(f"Successfully removed {val_path}")
        except PermissionError:
            print(f"Warning: Could not delete {val_path} due to Windows lock. "
                  "The folder is empty and the files are moved, so you can safely ignore this or delete it manually later.")

print("\nDone! All files should now be in the root 'images' and 'labels' folders.")
```

### Verification
After running the cell above, you should see `images/` and `labels/` folders inside `coco_reminder_set/` containing all 5,000 files.
