# P4: Balanced Merge and Split

**Goals:**
1.  Collect all data from `processed_data/`.
2.  **Oversample** smaller classes (like Stairs) to ensure balance.
3.  Integrate the **5,000 COCO Reminder Set**.
4.  Apply a global **70/15/15 Train/Val/Test Split**.
5.  Generate the final **84-class `data.yaml`** with correct mapping.

```python
import os
import shutil
import random
import yaml
from pathlib import Path

# --- CONFIG ---
BASE_DIR      = Path(r"c:\SafeStride\Dataset preprocessing")
PROCESSED_ROOT = BASE_DIR / "processed_data"
FINAL_DIR      = BASE_DIR / "final_dataset"
COCO_DIR       = BASE_DIR / "coco_reminder_set"

# OVERSAMPLING WEIGHTS
WEIGHTS = {
    "p1_pothole": 1,
    "p2_manhole": 1,
    "p3_staircase": 5,
    "coco_reminder_set": 1
}

def setup_final_dirs():
    # On Windows, rmtree often fails if a folder is open in Explorer.
    # We use ignore_errors=True to clear what we can and then ensure dirs exist.
    if FINAL_DIR.exists():
        shutil.rmtree(FINAL_DIR, ignore_errors=True)
    
    # Re-create structure
    for split in ["train", "val", "test"]:
        (FINAL_DIR / split / "images").mkdir(parents=True, exist_ok=True)
        (FINAL_DIR / split / "labels").mkdir(parents=True, exist_ok=True)

setup_final_dirs()
print(f"Final Dataset Path: {FINAL_DIR} (Folders cleared/prepared)")
```

## 1 — Collect and Weight Data Samples

```python
all_samples = []

def collect_from_dir(source_dir, weight):
    img_dir = source_dir / "images"
    lbl_dir = source_dir / "labels"
    count = 0
    
    if not lbl_dir.exists(): return 0
    
    for lf in lbl_dir.glob("*.txt"):
        img_file = None
        for ext in [".jpg", ".jpeg", ".png", ".bmp"]:
            candidate = img_dir / f"{lf.stem}{ext}"
            if candidate.exists():
                img_file = candidate
                break
        
        if img_file:
            for i in range(weight):
                all_samples.append({"img": img_file, "lbl": lf, "source": source_dir.name, "copy_idx": i})
            count += 1
    return count

# 1. Collect Hazard Data
for folder in PROCESSED_ROOT.iterdir():
    if not folder.is_dir(): continue
    w = WEIGHTS.get(folder.name, 1)
    c = collect_from_dir(folder, w)
    print(f"Source {folder.name}: Collected {c} images (Weight {w}x -> {c*w} total samples)")

# 2. Collect COCO Reminder Data
if COCO_DIR.exists():
    w = WEIGHTS.get("coco_reminder_set", 1)
    c = collect_from_dir(COCO_DIR, w)
    print(f"Source COCO: Collected {c} images (Weight {w}x -> {c*w} total samples)")
else:
    print("[WARNING] coco_reminder_set folder not found!")

print(f"\nTotal combined samples in pool: {len(all_samples)}")
```

## 2 — Global Shuffle and Split

```python
random.shuffle(all_samples)

# 70/15/15 Split
total = len(all_samples)
train_end = int(total * 0.7)
val_end   = int(total * 0.85)

train_set = all_samples[:train_end]
val_set   = all_samples[train_end:val_end]
test_set  = all_samples[val_end:]

def copy_to_final(subset, split_name):
    print(f"Copying {len(subset)} samples to {split_name}...")
    for item in subset:
        prefix = f"w{item['copy_idx']}_" if item['copy_idx'] > 0 else ""
        new_img_name = f"{prefix}{item['img'].name}"
        new_lbl_name = f"{prefix}{item['lbl'].name}"
        
        shutil.copy2(item["img"], FINAL_DIR / split_name / "images" / new_img_name)
        shutil.copy2(item["lbl"], FINAL_DIR / split_name / "labels" / new_lbl_name)

copy_to_final(train_set, "train")
copy_to_final(val_set, "val")
copy_to_final(test_set, "test")

print("\nBalanced Dataset Split Complete (70/15/15)!")
```

## 3 — Generate Final `data.yaml`

```python
# 1. Read the exported COCO mapping to fix the scrambled IDs
coco_yaml_path = COCO_DIR / "dataset.yaml"
if coco_yaml_path.exists():
    with open(coco_yaml_path, 'r') as f:
        coco_data = yaml.safe_load(f)
    # Get names in index order (0 to 79)
    coco_names_list = [coco_data['names'][i] for i in range(80)]
else:
    print("[ERROR] COCO dataset.yaml not found at", coco_yaml_path)
    coco_names_list = ['person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus', 'train', 'truck', 'boat', 'traffic light', 'fire hydrant', 'stop sign', 'parking meter', 'bench', 'bird', 'cat', 'dog', 'horse', 'sheep', 'cow', 'elephant', 'bear', 'zebra', 'giraffe', 'backpack', 'umbrella', 'handbag', 'tie', 'suitcase', 'frisbee', 'skis', 'snowboard', 'sports ball', 'kite', 'baseball bat', 'baseball glove', 'skateboard', 'surfboard', 'tennis racket', 'bottle', 'wine glass', 'cup', 'fork', 'knife', 'spoon', 'bowl', 'banana', 'apple', 'sandwich', 'orange', 'broccoli', 'carrot', 'hot dog', 'pizza', 'donut', 'cake', 'chair', 'couch', 'potted plant', 'bed', 'dining table', 'toilet', 'tv', 'laptop', 'mouse', 'remote', 'keyboard', 'cell phone', 'microwave', 'oven', 'toaster', 'sink', 'refrigerator', 'book', 'clock', 'vase', 'scissors', 'teddy bear', 'hair drier', 'toothbrush']

# 2. Add our 4 custom classes
final_names = coco_names_list + ['Pothole', 'Road Crack', 'Open Manhole', 'Stairs']

yaml_content = f"""
train: {FINAL_DIR}/train/images
val: {FINAL_DIR}/val/images
test: {FINAL_DIR}/test/images

nc: 84
names: {final_names}
"""

yaml_file = BASE_DIR / "final_training_config.yaml"
yaml_file.write_text(yaml_content.strip())
print(f"Generated 84-class config at: {yaml_file}")
```
